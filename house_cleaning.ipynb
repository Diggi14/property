```python
df=pd.read_csv('/content/property/houses.csv')
```

```python
df.head(5)
```

```python
df.drop(['link','property_id'],axis=1,inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['society'].value_counts()
```

```python
def remove_rating(x):
  x=str(x)
  if('★' in x):
    return x[:-5]
  else:
    return x
```

```python
df['society']=df['society'].apply(remove_rating)
```

```python
df['society'].value_counts()
```

```python
df['society']=df['society'].str.replace('nan','independent')
```

```python
df.head(5)
```

```python
df['price']
```

```python
df=df[df['price']!='Price on Request']
```

```python
df['price'].isnull().sum()
```

```python
df.dropna(subset=['price'],inplace=True)
```

```python
def price_convert(x):
  x=str(x)
  number,amount=x.split(' ')
  if(amount=='Lac'):
    return float(number)/100
  elif(amount=='Crore'):
    return float(number)
```

```python
df['price']=df['price'].apply(price_convert)
df['price'] = df['price'].round(2)
```

```python
df.head(5)
```

```python
df['rate'].value_counts()
```

```python
df['rate'].isnull().sum()
```

```python
df.rename(columns={'rate':'price_per_sqft'},inplace=True)
```

```python
df['rate']=df['rate'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df['price_per_sqft']=df['price_per_sqft'].astype(int)
```

```python
df['area']=df['area'].str.split('sq').str.get(0).str.split('(').str.get(1)
```

```python
df.head(5)
```

```python
df['bedRoom'].isnull().sum()
```

```python
df['bedRoom']=df['bedRoom'].str.split(' ').str.get(0)
```

```python
df['bedRoom']=df['bedRoom'].astype(int)
```

```python
df['bathroom'].value_counts()
```

```python
df['bathroom']=df['bathroom'].str.split(' ').str.get(0)
df['bathroom']=df['bathroom'].astype(int)
```

```python
df['balcony'].value_counts()
```

```python
df['balcony']=df['balcony'].str.split(' ').str.get(0)
```

```python
df['balcony']=df['balcony'].str.replace('No','0')
```

```python
df['additionalRoom'].value_counts()
```

```python
df['additionalRoom'].isnull().sum()
```

```python
df['additionalRoom'] = df['additionalRoom'].fillna('Not available')
```

```python
df.head(5)
```

```python
df.rename(columns={'noOfFloor':'floorNum'},inplace=True)
```

```python
df['floorNum'].isnull().sum()
```

```python
df[df['floorNum'].isnull()]
```

```python
df['floorNum'] = df['floorNum'].str.split(' ').str.get(0)
```

```python
df.to_csv('cleaned_flat.csv',index=False)
```
