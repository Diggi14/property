```python
df = df.drop_duplicates()
```

```python
df['society'].value_counts()
```

```python
df.dropna(subset=['society'],inplace=True)
```

```python
df[df['society'].isna()]
```

```python
import re
```

```python
def remove_rating(x):
  x=str(x)
  if('★' in x):
    return x[:-5]
  else:
    return x
```

```python
df['society']=df['society'].apply(remove_rating)
```

```python
df['society'].value_counts()
```

```python
df.drop(['link','property_id'],axis=1,inplace=True)
```

```python
df.sample(5)
```

```python
df['price']
```

```python
df=df[df['price']!='Price on Request']
```

```python
df[df['price']=='Price on Request']
```

```python
df['price'].isna().sum()
```

```python
df.dropna(subset=['price'],inplace=True)
```

```python
df.head(5)
```

```python
def price_convert(x):
  x=str(x)
  number,amount=x.split(' ')
  if(amount=='Lac'):
    return float(number)/100
  elif(amount=='Crore'):
    return float(number)
```

```python
df['price']=df['price'].apply(price_convert)
df['price'] = df['price'].round(2)
```

```python
df.head(5)
```

```python
df.info()
```

```python
df['area'].value_counts()
```

```python
df['area'].isnull().sum()
```

```python
df['area']=df['area'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','')
```

```python
df.rename({'area':'price_per_sq.ft'},inplace=True,axis=1)
```

```python
df.head(5)
```

```python
df['bedRoom'].value_counts()
```

```python
df['bedRoom']=df['bedRoom'].str.split(' ').str.get(0)
```

```python
df.info()
```

```python
df['price_per_sq.ft']=df['price_per_sq.ft'].astype(int)
df['bedRoom']=df['bedRoom'].astype(int)
```

```python
df.head(5)
```

```python
df['bathroom'].value_counts()
```

```python
df['bathroom']=df['bathroom'].str.split(' ').str.get(0)
df['bathroom']=df['bathroom'].astype(int)
```

```python
df['balcony'].value_counts()
```

```python
df['balcony']=df['balcony'].str.split(' ').str.get(0)
```

```python
df['balcony']=df['balcony'].str.replace('No','0')
```

```python
df['additionalRoom'].value_counts()
```

```python
df['additionalRoom'].isnull().sum()
```

```python
df['additionalRoom'] = df['additionalRoom'].fillna('Not available')
```

```python
df['floorNum'].isnull().sum()
```

```python
df[df['floorNum'].isnull()]
```

```python
df['floorNum'].fillna('6th of x',inplace=True)
```

```python
df['floorNum'].value_counts()
```

```python
import re
```

```python
df['floorNum']=df['floorNum'].str.split('of').str.get(0).str.replace('Ground','0s').str.replace('Basement','-999s').str.replace('Lower Ground','0s').str.replace(r'\D+', '', regex=True)
```

```python
df['floorNum']=df['floorNum'].str.replace('999','-1').astype(int)
```
