```python

popular_sectors = df_combined['sector'].value_counts()[lambda x: x >= 3].index

df = df_combined[df_combined['sector'].isin(popular_sectors)].copy()
```

```python
df['sector'] = df['sector'].str.strip()
df.loc[df['sector'].str.contains('n/a'),'sector']=df.loc[df['sector'].str.contains('n/a'),'sector'].str.replace('n/a','')
df.loc[df['sector'].str.contains('0001'),'sector']=df.loc[df['sector'].str.contains('0001'),'sector'].str.replace('0001','')
df.loc[df['sector'].str.contains('12th floor '),'sector']=df.loc[df['sector'].str.contains('12th floor '),'sector'].str.replace('12th floor ','')
df.loc[df['sector'].str.contains('antriksh heights '),'sector']=df.loc[df['sector'].str.contains('antriksh heights '),'sector'].str.replace('antriksh heights ','')
df.loc[df['sector'].str.contains('middle floor'),'sector']=df.loc[df['sector'].str.contains('middle floor'),'sector'].str.replace('middle floor','')
```

```python
df.loc[df['sector'].str.contains('a123'),'sector']=df.loc[df['sector'].str.contains('a123'),'sector'].str.replace('a123','')
```

```python
df['sector'] = (
    df['sector']
    .str.lower()
    .str.strip()
    .str.replace(r'\b(sector\s+\w+)\s+\1\b', r'\1', regex=True)
)
```

```python
df['sector'] = (
    df['sector']
    .str.lower()
    .str.strip()
    .str.replace(r'sector[-\s]*81\s+sector\s*81', 'sector 81', regex=True)
)
```

```python
df['sector'] = (
    df['sector']
    .str.lower()
    .str.strip()
    .str.replace(r'^sohna\s+', '', regex=True)
)
```
